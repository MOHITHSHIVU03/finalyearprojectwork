```python
#to identify the medications provided to the patient
import spacy
med7 = spacy.load('en_core_med7_lg')
# create distinct colours for labels
col_dict = {}
seven_colours = ['#e6194B', '#ba88a6', '#b88c12', '#ffd8b1', '#f58231', '#bec4eb', '#42d4f4']
for label, colour in zip(med7.pipe_labels['ner'], seven_colours):
    col_dict[label] = colour

options = {'ents': med7.pipe_labels['ner'], 'colors':col_dict}
```

```python
#file preprocessing
text=''
with open(r"C:\Users\Kavya\Downloads\train.test.released.8.17.09.tar\train.test.released.8.17.09\train.test.released.8.17.09\119940","r") as a_file:
    for line in a_file:
        stripped_line = line.replace("\n"," ")
        text+=stripped_line
#text
```
